# Indexing 

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
olymbic_df = pd.read_csv('./data/olympics.csv',skiprows=4)
olymbic_df.head()
```

### Index are immutable, we can assign a value to index

```python
type(olymbic_df.index)
```

```python
olymbic_df.index
```

### Default Index is integer based, we can change index as column

## set_index()

```python
olymbic_df.set_index('Athlete')
```

## To set the index in same data frame object do 'inplace=True'

```python
olymbic_df.set_index('Athlete',inplace=True)
olymbic_df.head()
```

```python
olymbic_df.reset_index(inplace=True)
olymbic_df.head()
```

## sort_index()

```python
olymbic_df.set_index('Athlete',inplace=True)
olymbic_df.sort_index(inplace=True,ascending=False)
olymbic_df.head()
```

## loc[...]

```python
olymbic_df.loc['BOLT, Usain']
```

```python
olymbic_df.reset_index(inplace=True)
```

```python
olymbic_df.head()
```

```python
olymbic_df.loc[olymbic_df.Athlete == 'BOLT, Usain']
```

## iloc[...]
- Used for python list slicing 

```python
olymbic_df.iloc[1700]
```

```python
olymbic_df.iloc[[1542, 2390,6000,15000]]
```

```python
olymbic_df.iloc[1:4]
```

### Plot the total number of medals awarded at each of the Olympic games throughout history.

```python
olymbic_df.Edition.value_counts().sort_index().plot()
```
